```python
# load the needed modules, follow pandas method outlined in the readme hint
import pandas as pd
import os
import csv
```

```python
# Access the file path
csv_path = os.path.join("cities.csv")
```

```python
# Assign file to df per instructions in the hint, check top 20 rows
city_file_df = pd.DataFrame(city_file)
city_file_df.head(20)
```

```python
# Data above looks good, just need to change date into a more standard form using datetime method
# Found method on stack overflow here that worked... https://stackoverflow.com/questions/42958217/pandas-read-excel-datetime-converter
import datetime
```

```python
city_file_df["Date"] = pd.to_datetime(city_file_df["Date"])
city_file_df.head(20)
```
